```python
mean_rating = df['Rating'].mean()
df['Rating']=df['Rating'].fillna(mean_rating)
df['Type']=df['Type'].fillna('NaN')
df['Content Rating']=df['Content Rating'].fillna('NaN')
df['Current Ver']=df['Current Ver'].fillna(0)
df['Android Ver']=df['Android Ver'].fillna(0)
```

```python
df.isnull().sum()
```

```python
a=df.apply(lambda col: col.unique())
print(a)
```

```python
df1=df.copy()
```

```python
df1[df1['Category']=='1.9']
```

```python
df1=df1.drop(10472,axis=0).reset_index()
```

1. Average Rating by Category

```python
df1['Category'].unique()
```

```python
category_rating = df1.groupby('Category')['Rating'].mean().sort_values()

plt.figure(figsize=(16, 8))
category_rating.plot(kind='bar', color='blue')
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.show()
```

 2.Number of installs varies across different app categories

```python
df1['Installs'] = df1['Installs'].str.replace('+', '')
```

```python
df1['Installs'] = df1['Installs'].str.replace(',', '')
```

```python
df1['Installs'] = df1['Installs'].str.replace('Free', '0')
```

```python
df1['Installs']=df1['Installs'].astype(int)
```

```python
df1['Installs'].unique()
```

```python
category_installs = df1.groupby('Category')['Installs'].mean().sort_values()

plt.figure(figsize=(16,4))
category_installs.plot(kind='bar', color='blue')
plt.title('Average Number of Installs by Category')
plt.xlabel('Category')
plt.ylabel('Average Number of Installs')
plt.show()
```

3.Average rating of free and paid apps

```python
df3= df1[df1['Type']!='NaN']
```

```python
b=df3.apply(lambda col: col.unique())
print(b)
```

```python
type_rating = df3.groupby('Type')['Rating'].mean()

plt.figure(figsize=(8, 5))
type_rating.plot(kind='bar', color=['blue','pink'])
plt.title('Average Rating of Free vs. Paid Apps')
plt.xlabel('App Type')
plt.ylabel('Average Rating')
plt.show()
```

4.Total price v/s Genres

```python
df1['Price'] = df1['Price'].str.replace('$', ' ')
```

```python
df1['Price']=df1['Price'].astype(float)
```

```python
df1['Price'].dtypes
```

```python
App_price = df1.groupby('Genres')['Price'].mean()

plt.figure(figsize=(24, 3))
App_price.plot(kind='line', color='blue')
plt.title('Total price v/s Genres ')
plt.xlabel('Genres')
plt.show()
```

5.Distribution of apps across different content ratings

```python
content_rating_distribution = df1.groupby('Content Rating')['App'].count()
plt.figure(figsize=(6,6))
content_rating_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Apps by Content Rating')
plt.ylabel('')
plt.show()
```

6.Number of reviews for apps varies over the years.

```python
df1['Reviews'].dtypes
df1['Reviews']=df1['Reviews'].astype('float')
```

```python
df1['Last Updated'] = pd.to_datetime(df1['Last Updated'])
df1['Year'] = df1['Last Updated'].dt.year
plt.figure(figsize=(10, 6))
plt.scatter(df1['Year'], df1['Reviews'], alpha=0.5, color='orange')
plt.title('Number of Reviews vs. Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.yscale('log')  # Use a logarithmic scale for reviews since they can vary greatly
plt.grid(True)
plt.show()
```

7.Month-Wise Average Rating Over Time

```python
df1['Month'] = df1['Last Updated'].dt.month
month_wise_rating = df1.groupby('Month')['Rating'].mean()

plt.figure(figsize=(9, 6))
month_wise_rating.plot(kind='line', color='blue', marker='o')
plt.title('Month-Wise Average Rating Over Time')
plt.xlabel('Month')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```
